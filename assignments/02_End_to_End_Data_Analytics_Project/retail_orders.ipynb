```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#renae columns to lower case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
# derive discount,sale price and profit columns
df['discount'] = df['list_price'] * df['discount_percent'] * .01
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
# conver date type from object to date
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# drop list_price, cost_price and discount_percent columns
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
#load the data into sql server using replace option
import sqlalchemy as sal
engine = sal.create_engine(
    "mssql+pyodbc://@Taha/retail_orders?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)
conn=engine.connect()
```
